```python
import pandas as pd

# Read datasets
original_data = pd.read_csv('data/cleaned_data_SYMPTOMS_9_13_23_DNA.csv')
dataset1 = pd.read_csv('data/dataset_1.csv').drop('Unnamed: 0', axis=1)
dataset2 = pd.read_csv('data/dataset_2.csv').drop('Unnamed: 0', axis=1)
dataset3 = pd.read_csv('data/dataset_3.csv').drop('Unnamed: 0', axis=1)
dataset4 = pd.read_csv('data/dataset_4.csv').drop('Unnamed: 0', axis=1)
dataset5 = pd.read_csv('data/dataset_5.csv').drop('Unnamed: 0', axis=1)

# Assuming you've already loaded your original data DataFrame
# original_data = pd.read_csv('path_to_original_data.csv')

def verify_dataset(dataset, original_data, name):
    print(f"Verification for {name}:")
    
    # Shape
    print(f"Shape: {dataset.shape}")
    
    # Columns added or removed
    added_columns = set(dataset.columns) - set(original_data.columns)
    removed_columns = set(original_data.columns) - set(dataset.columns)
    if added_columns:
        print(f"Added columns: {list(added_columns)}")
    if removed_columns:
        print(f"Removed columns: {list(removed_columns)}")
    
    # Check normalization of FAS
    if 'FAS' in dataset.columns:
        print(f"FAS range: {dataset['FAS'].min()} to {dataset['FAS'].max()}")
    
    # Check one-hot encoded columns
    one_hot_cols = [col for col in dataset.columns if '_' in col and col not in original_data.columns]
    for col in one_hot_cols:
        print(f"{col} unique values: {dataset[col].unique()}")
    
    # Verify if columns are binary or normalized
    non_binary_cols = []
    for col in dataset.columns:
        unique_vals = dataset[col].unique()
        if set(unique_vals) != {0, 1}:
            try:
                mean_val = dataset[col].mean()
                print(f"Column '{col}' is not binary. Mean value: {mean_val}")
            except TypeError:
                print(f"Column '{col}' has mixed data types. Unique values: {unique_vals}")
    
    print("\n" + "-"*50 + "\n")

# Verify each dataset
verify_dataset(dataset1, original_data, "Dataset 1")
verify_dataset(dataset2, original_data, "Dataset 2")
verify_dataset(dataset3, original_data, "Dataset 3")
verify_dataset(dataset4, original_data, "Dataset 4")
verify_dataset(dataset5, original_data, "Dataset 5")
```

```python
def display_fas_cat_dummies(df):
    fas_cat_columns = [col for col in df.columns if 'Flag' in col]
    
    if fas_cat_columns:
        print("FAS_cat dummy variables present in the dataframe:")
        for col in fas_cat_columns:
            print(col)
            print(df[col].value_counts())
            print('-' * 50)
    else:
        print("No FAS_cat dummy variables found in the dataframe.")

# Example usage:
display_fas_cat_dummies(dataset3)
```
